```python
from river import datasets
import pandas as pd
from fluvialgen.movingwindow_generator import MovingWindowBatcher
from collections import Counter
```

```python
nums = [0, 1, 2,3, 4]
def num_list_to_river(nums):
    for n in nums:
        yield n

river_stream = num_list_to_river(nums)
```

```python
#dataset = datasets.Bikes()

batcher = MovingWindowBatcher(
    dataset=river_stream,
    instance_size=5,
    batch_size=7,
    n_instances=100
)
```

```python
x_batches = []
y_batches = []

for x, y in batcher:
    x_batches.append(x)
    y_batches.append(y)
```

```python
df_merged = pd.concat(x_batches, ignore_index=True)
df_merged
```

```python
conteo_duplicados = df_merged.groupby(list(df_merged.columns)).size().reset_index(name='Count')
```
